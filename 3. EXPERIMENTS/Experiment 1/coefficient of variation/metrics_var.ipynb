```python
import pandas as pd
```

```python
def compute_cv(group1, group2, group3, group4, group5, group6, group7, group8):
    # Calculate the mean and standard deviation for each group
    means = [group.mean() for group in [group1, group2, group3, group4, group5, group6, group7, group8]]
    stds = [group.std() for group in [group1, group2, group3, group4, group5, group6, group7, group8]]
    
    # Calculate CV for each group
    cv1 = (stds[0] / means[0])  if means[0] != 0 else float('nan')
    cv2 = (stds[1] / means[1]) if means[1] != 0 else float('nan')
    cv3 = (stds[2] / means[2])  if means[2] != 0 else float('nan')
    cv4 = (stds[3] / means[3])  if means[3] != 0 else float('nan')
    cv5 = (stds[4] / means[4])  if means[4] != 0 else float('nan')
    cv6 = (stds[5] / means[5])  if means[5] != 0 else float('nan')
    cv7 = (stds[6] / means[6])  if means[6] != 0 else float('nan')
    cv8 = (stds[7] / means[7])  if means[7] != 0 else float('nan')

    # Store CV values
    cvs = [cv1, cv2, cv3, cv4, cv5, cv6, cv7, cv8]

    # Calculate the mean CV of the groups
    mean_cv = sum(cvs) / len(cvs)

    # Combine all groups into a single series
    combined_data = pd.concat([group1, group2, group3, group4, group5, group6, group7, group8])

    # Calculate the mean and standard deviation of the combined data
    combined_mean = combined_data.mean()
    combined_std = combined_data.std()

    # Calculate the combined CV
    combined_cv = (combined_std / combined_mean) if combined_mean != 0 else float('nan')

    # Create a DataFrame with the CVs and combined CV
    df = pd.DataFrame({
        'CV Qasper Aligned': [cv1],
        'CV Qasper Shuffled': [cv2],
        'CV Qaper 8b': [cv3],
        'CV Qasper 70b': [cv4],
        'CV Qasa Aligned': [cv5],
        'CV Qasa Shuffled': [cv6],
        'CV Qasa 8b': [cv7],
        'CV Qasa 70b': [cv8],
        'Combined CV': [combined_cv],
        'Mean CV': [mean_cv]
    })

    return df
```
