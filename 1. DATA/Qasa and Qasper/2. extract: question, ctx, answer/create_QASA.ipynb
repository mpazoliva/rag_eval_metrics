## QASA DATASET

```python
import json
import pandas as pd
```

```python
# Load and read .json file as pandas df
json_file_path = '/home/ubuntu/iris_repos/llm_evaluation_thesis/QASA-QASPER/1. original_data/qasa_data.json'  

with open(json_file_path, 'r') as f:
    data = json.load(f)

flat_data = [data[key] for key in data]
df = pd.json_normalize(flat_data)
```

```python
df.head()
```

```python
# Remove unnecessary columns and rename some
columns_to_drop = ['paper_id', 'title', 'question_id', 'question_section',
       'question_trigger_sentence', 'arxiv_id', 's2orc_url', 'arxiv_url', 's2orc_id']
df = df.drop(columns=columns_to_drop)

new_column_names = {'composition': 'correct_answer', 'evidential_info': 'context'}
df.rename(columns=new_column_names, inplace=True)
```

```python
df
```

```python
# Extract context to a list of strings instead of a dictionary
pd.set_option('display.max_rows', None)    
pd.set_option('display.max_columns', None)
print(df["context"].head())
df['context'] = df['context'].apply(lambda x: [entry['context'] for entry in x] if isinstance(x, list) else None)
print(df["context"].head())
```

```python
df.head()
```

```python
types = df['context'].apply(type)
unique_types = types.unique()
print(unique_types)
```

```python
list_lengths = df['context'].apply(len)
print(list_lengths)
```

```python
# Extract context as a single string
merge_strings = lambda x: ' '.join(x)
df['context'] = df['context'].apply(merge_strings)
```

```python
df.head()
```

```python
# Unify the values in question type
df['question_type'] = df['question_type'].str.lower()
```

```python
df = df[(df['question_type'] != '-') & (df['question_type'] != 'shallow/testing')]
```

```python
value_counts = df['question_type'].value_counts()
print(value_counts)
```

```python
# Map old values to new values
mapping = {
    'deep/complex question': 'complex question',
    'deep/complex question ': 'complex question',
    'testing': 'testing question',
    'testing ': 'testing question',
    'testing question ': 'testing question',
    'shallow': 'shallow question',
    'shallow question ': 'shallow question',
    'deep / complex': 'complex question',
    'complex': 'complex question',
    'deep / complex question ': 'complex question',
    'deep / complex question': 'complex question',
    'deep question': 'complex question',
}

# Replace values in the "question_type" column
df['question_type'] = df['question_type'].replace(mapping)
```

```python
df = df[(df['question_type'] != 'deep question')]
```

```python
value_counts = df['question_type'].value_counts()
print(value_counts)
```

```python
df.columns
```

```python
# Remove rows with empty context
df = df[df['context'].astype(str).str.strip() != '']
```

```python
# INFORMATION OF QUESTION COLUMN (by characters)

# Calculate statistics for string length in a column
string_lengths = df['question'].apply(lambda x: len(str(x)))

# Compute statistics
average_length = string_lengths.mean()
min_length = string_lengths.min()
max_length = string_lengths.max()
median_length = string_lengths.median()
std_dev = string_lengths.std()

# Print statistics
print("Average Length:", average_length)
print("Minimum Length:", min_length)
print("Maximum Length:", max_length)
print("Median Length:", median_length)
print("Standard Deviation:", std_dev)
```

```python
# INFORMATION OF CONTEXT COLUMN (by characters)

# Calculate statistics for string length in a column
string_lengths = df['context'].apply(lambda x: len(str(x)))

# Compute statistics
average_length = string_lengths.mean()
min_length = string_lengths.min()
max_length = string_lengths.max()
median_length = string_lengths.median()
std_dev = string_lengths.std()

# Print statistics
print("Average Length:", average_length)
print("Minimum Length:", min_length)
print("Maximum Length:", max_length)
print("Median Length:", median_length)
print("Standard Deviation:", std_dev)
```

```python
# INFORMATION OF ANSWER COLUMN (by characters)

# Calculate statistics for string length in a column
string_lengths = df['correct_answer'].apply(lambda x: len(str(x)))

# Compute statistics
average_length = string_lengths.mean()
min_length = string_lengths.min()
max_length = string_lengths.max()
median_length = string_lengths.median()
std_dev = string_lengths.std()

# Print statistics
print("Average Length:", average_length)
print("Minimum Length:", min_length)
print("Maximum Length:", max_length)
print("Median Length:", median_length)
print("Standard Deviation:", std_dev)
```
