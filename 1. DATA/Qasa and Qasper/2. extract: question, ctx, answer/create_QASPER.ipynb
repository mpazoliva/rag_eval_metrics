```python
import json
import pandas as pd
```

```python
json_files = ['/home/ubuntu/iris_repos/data_preparation(QASA-QASPER)/1. original_data/qasper1.json', '/home/ubuntu/iris_repos/data_preparation(QASA-QASPER)/1. original_data/qasper2.json']  
```

```python
# Initialize lists to store data
questions = []
answers_list = []
contexts_list = []

# Iterate through each JSON file
for json_file in json_files:
    # Load JSON data from the file
    with open(json_file, 'r') as file:
        data = json.load(file)

    # Iterate through each instance in the JSON data
    for key, instance in data.items():
        # Iterate through each question in the 'qas' section
        for qas_instance in instance.get('qas', []):
            # Extract question, answers, and contexts
            question = qas_instance.get('question', '')
            answers = [answer_instance['answer']['free_form_answer'] for answer_instance in qas_instance.get('answers', [])]
            contexts = [evidence_instance for answer_instance in qas_instance.get('answers', []) for evidence_instance in answer_instance['answer']['evidence']]

            # Append to lists only if 'answers' is not empty
            if any(answers):
                questions.append(question)
                answers_list.append(answers)
                contexts_list.append(contexts)

# Create a DataFrame
df = pd.DataFrame({'question': questions, 'correct_answer': answers_list, 'context': contexts_list})
```

```python
# Extract context as a single string
merge_strings = lambda x: ' '.join(x)
df['context'] = df['context'].apply(merge_strings)
df['correct_answer'] = df['correct_answer'].apply(merge_strings)
```

```python
# INFORMATION OF QUESTION COLUMN (by characters)

# Calculate statistics for string length in a column
string_lengths = df['question'].apply(lambda x: len(str(x)))

# Compute statistics
average_length = string_lengths.mean()
min_length = string_lengths.min()
max_length = string_lengths.max()
median_length = string_lengths.median()
std_dev = string_lengths.std()

# Print statistics
print("Average Length:", average_length)
print("Minimum Length:", min_length)
print("Maximum Length:", max_length)
print("Median Length:", median_length)
print("Standard Deviation:", std_dev)
```
