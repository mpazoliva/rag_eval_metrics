```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to plot and save bar chart for a DataFrame
def plot_bar_chart(df, metric_name):
    # Calculate means for each required column
    means = {
        'perfect': df[f'{metric_name}_ground_truth'].mean(),
        'c_similar': df[f'{metric_name}_correct_similar'].mean(),
        'different': df[f'{metric_name}_correct_different'].mean(),
        'i_similar': df[f'{metric_name}_incorrect_similar'].mean(),
        'related': df[f'{metric_name}_incorrect_related'].mean(),
        'unrelated': df[f'{metric_name}_incorrect_unrelated'].mean()
    }
    
    # Extract the labels and values for the plot
    labels = list(means.keys())
    values = list(means.values())
    
    # Define colors for the bars
    colors = ['#d4f0d2', '#d4f0d2', '#d4f0d2', '#f2d1d1', '#f2d1d1', '#f2d1d1']
    
    # Create the bar plot
    plt.figure(figsize=(10, 5))
    plt.bar(labels, values, color=colors)
    plt.xlabel('Categories')
    plt.ylabel('Mean Value')
    
    # Save the plot to a file
    save_filename = f'{metric_name}_bar_chart.png'  # Define file name
    plt.savefig(f'{save_filename}', bbox_inches='tight')  # Save plot with tight bounding box
    
    plt.show()  # Display the plot if needed
    
    print(f'Chart saved as {save_filename}')  # Confirmation message

```

```python
plot_bar_chart(bleu, 'bleu')  
```
